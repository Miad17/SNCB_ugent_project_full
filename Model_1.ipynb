```python
# read in punctuality data
punct2019_1 = pd.read_csv("./data/Punctuality_2019_part1.csv", sep = ";")
punct2019_2 = pd.read_csv("./data/Punctuality_2019_part2.csv", sep = ";")
punct2020_1 = pd.read_csv("./data/Punctuality_2020_part1.csv", sep = ";")
punct2020_2 = pd.read_csv("./data/Punctuality_2020_part2.csv", sep = ";")
punct2021_1 = pd.read_csv("./data/Punctuality_2021_part1.csv", sep = ";")
punct2021_2 = pd.read_csv("./data/Punctuality_2021_part2.csv", sep = ";")
punct2022_1 = pd.read_csv("./data/Punctuality_2022_part1.csv", sep = ";")
punct2022_2 = pd.read_csv("./data/Punctuality_2022_part2.csv", sep = ";")
punct2023 = pd.read_csv("./data/Punctuality_2023.csv", sep = ";")
# combine all punctuality
all_punct = pd.concat([punct2019_1, punct2019_2, punct2020_1, punct2020_2, punct2021_1, punct2021_2, punct2022_1, punct2022_2, punct2023], axis = 0)
# create year var
all_punct['DAT_DEP'] = pd.to_datetime(all_punct['DAT_DEP'])
all_punct['YEAR_DEP'] = all_punct['DAT_DEP'].apply(lambda x: x.year)
# remove rows with NaN TRAIN_ID
all_punct = all_punct[all_punct['TRAIN_ID'].notna()]

# read in justifications data
justifs = pd.read_csv("/Users/miek/Downloads/Studie P-treinen/Extract justifications.csv", sep = ";", encoding='latin-1')
justifs['p_resp'] = justifs['trno_resp'].apply(lambda x: 1 if ((x >= 7000) & (x <= 8999)) else 0)

# count how many times a train was justified
trains_justifs_count = justifs.groupby(['REF_TOD_IMPACTED_TRAIN']).agg(justif_count=('REF_JUS_ID', 'count'),
                                                                       p_resp_count=('p_resp', 'sum')).reset_index()
punct_justifs = all_punct.merge(trains_justifs_count, left_on = 'TRAIN_ID', right_on='REF_TOD_IMPACTED_TRAIN', how = "left")
punct_justifs = punct_justifs.drop(columns='REF_TOD_IMPACTED_TRAIN')
punct_justifs['justif_count'].fillna(0, inplace=True)
punct_justifs['p_resp_count'].fillna(0, inplace=True)
punct_justifs

```

```python
# read tjj_ptcar_parquet
stops = pd.read_parquet("~/Downloads/Data P-trein studie (20042023)/TJJ_PTCAR_parquet")
stops

# read ptcar.csv
ptcar = pd.read_csv("./data/PTCAR.csv", sep = ";")
ptcar

# join stops and ptcar
stops = stops.merge(ptcar, on=['REF_PTCAR_NO'], how = "left")

# keep only commercial stops for one-hot-encoding purposes
commercial_stops = stops[stops['COMMERCIAL_STOP_NATURE'] > 0].reset_index().drop(columns="index")

# add indicator if train was delayed for 90 sec or more at a commercial stop
commercial_stops['small_delay'] = commercial_stops['EH_ARR'].apply(lambda x: 1 if x >= 90 else 0)

# group by REF_TOD_TRAIN_ID, aggregate and small delays
commercial_stops_grouped = commercial_stops.groupby('REF_TOD_TRAIN_ID').agg(sum_small_delay=('small_delay', 'sum'),
                                                 count_small_delay=('small_delay', 'size'))\
                                                 .reset_index()

commercial_stops_grouped['small_delay_prop'] = commercial_stops_grouped['sum_small_delay'] / commercial_stops_grouped['count_small_delay']
commercial_stops_grouped = commercial_stops_grouped.drop(columns=['sum_small_delay', 'count_small_delay'])


# add one-hot for stations stopped at by each REF_TOD_TRAIN_ID
# one-hot-encode LONG_NAME_NL column
one_hot_encoded = pd.get_dummies(commercial_stops['LONG_NAME_NL'])

# concatenate one-hot-encoded columns with original dataframe
commercial_stops_ohe = pd.concat([commercial_stops, one_hot_encoded], axis=1)

# group by REF_TOD_TRAIN_ID and get the maximum value of each station
grouped_stops = commercial_stops_ohe.groupby('REF_TOD_TRAIN_ID')[one_hot_encoded.columns.tolist()].max()

# reset the index and rename the columns
grouped_stops = grouped_stops.reset_index()

# join grouped_stops columns with commercial_stops_grouped
commercial_stops_grouped = commercial_stops_grouped.merge(grouped_stops, on='REF_TOD_TRAIN_ID')

# create column to signal if train was in the morning rush or evening rush
first_stops = stops[stops['PTCAR_ORD_NO']==101]
def get_rush(hour):
    if 6 <= hour < 9:
        return 'morning_rush'
    elif 16 <= hour < 19:
        return 'evening_rush'
    else:
        return 'no_rush'
first_stops['rush'] = first_stops['THR_DEP'].apply(lambda x: get_rush(x.hour)).astype('category')

# join commercial_stops_grouped with the rush from first_stops_df
commercial_stops_grouped = commercial_stops_grouped.merge(first_stops[['REF_TOD_TRAIN_ID', 'rush']], on='REF_TOD_TRAIN_ID')

# check
commercial_stops_grouped
```
