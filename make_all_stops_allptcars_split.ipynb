```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import seaborn as sns
import os
import holidays
import datetime as dt
os.getcwd()
```

```python
stops = pd.read_parquet("~/Downloads/all_stops_ungrouped/all_stops_ungrouped.parquet")
stops
```

```python
stops.columns
```

```python
# Create DataFrames for each time period
stops_pre_covid = stops[stops['date_origin'] < pd.to_datetime('2020-03-18')].reset_index()
stops_pre_covid = stops_pre_covid.drop(columns='index')

stops_during_covid = stops[(stops['date_origin'] >= pd.to_datetime('2020-03-18')) & (stops['date_origin'] <= pd.to_datetime('2022-05-11'))].reset_index()
stops_during_covid = stops_during_covid.drop(columns='index')

stops_post_covid = stops[stops['date_origin'] > pd.to_datetime('2022-05-11')].reset_index()
stops_post_covid = stops_post_covid.drop(columns='index')
```

```python
# make parquets
stops_pre_covid.to_parquet("~/Downloads/all_stops_ungrouped/all_stops_pre_covid.parquet")
stops_during_covid.to_parquet("~/Downloads/all_stops_ungrouped/all_stops_during_covid.parquet")
stops_post_covid.to_parquet("~/Downloads/all_stops_ungrouped/all_stops_post_covid.parquet")
```
